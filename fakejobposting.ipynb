## NLP

```python
df_lower=df_s['combined_text'].str.lower()
```

```python
df_lower.loc[0]
```

```python
df_lower.loc[1]
```

```python
import re

def remove_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub('', text)
```

```python
df_no_url = df_lower.apply(remove_url)
```

```python
df_no_url.loc[6]
```

```python
import string
string.punctuation
exclude=string.punctuation
```

```python
import string
def clean_text(text):
    text = text.replace('\xa0', ' ')
    text = text.replace('’', "'").replace('‘', "'")
    text = text.replace('“', '"').replace('”', '"')
    for char in exclude:
        text = text.replace(char, '')
    text = re.sub(r'\s+', ' ', text)
    return text.strip()
```

```python
df_clean_final = df_no_url.apply(clean_text)
```

```python
df_clean_final.loc[6]
```

```python
df_clean_final.loc[0]
```

```python
from nltk.corpus import stopwords
stop = set(stopwords.words('english'))
```

```python
def remove_stopwords(text):
    if not isinstance(text, str):
        return text   # NaN ya non-string ko as-it-is chhod do
    
    words = []
    for word in text.split():
        if word not in stop:
            words.append(word)
    
    return " ".join(words)
```

```python
df_no_stopword = df_clean_final.apply(remove_stopwords)
```

```python
df_no_stopword.loc[0]
```

## now, modelling using TF-IDF 

```python
x=df_no_stopword
y=df['fraudulent']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=42, stratify=y)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(max_features=5000,
                      ngram_range=(1,2),
                     )
x_train_tfidf=tfidf.fit_transform(x_train)
x_test_tfidf=tfidf.transform(x_test)
```

## logistic regression

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(
    max_iter=1000,
    class_weight={0:1, 1:2}
)

model.fit(x_train_tfidf, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = model.predict(x_test_tfidf)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

## svm

```python
from sklearn.svm import LinearSVC

model = LinearSVC(
    class_weight={0:1, 1:2}
)

model.fit(x_train_tfidf, y_train)
```

```python
y_pred = model.predict(x_test_tfidf)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

print("Accuracy:", accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(x_test_tfidf)

print(classification_report(y_test, y_pred))
```
